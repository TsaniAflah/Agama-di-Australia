```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_can = pd.read_csv('Australian religion 1881 - 2016.csv')
```

*NOTE : df_can dapat diganti dengan sebutan lainnya*

```python
df_can.head()
```

```python
print(df_can.shape)
```

```python
df_can.drop(["1881", "1882", "1883", "1884", "1885", "1886", "1887", "1888", "1889", "1890"], axis=1, inplace=True)
```

```python
df_can.head()
```

```python
df_can.rename(columns={'Religion':'Belief'}, inplace=True)

df_can.head()
```

```python
all(isinstance(column, str) for column in df_can.columns)
```

```python
df_can.set_index('Belief', inplace=True)

df_can.head()
```

```python
df_can.drop(["Uniting Church"], axis=0, inplace=True)
```

```python
df_can.head()
```

```python
df_can['Total'] = df_can.sum(axis=1)

df_can.head()
```

```python
print('data dimensions:', df_can.shape)
```

```python
years = list(map(str, range(1891, 2017)))
years
```

**AREA PLOT**

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')  # optional: for ggplot-like style

# check for latest version of Matplotlib
print('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5 = df_can.head()

# transpose the dataframe
df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',
             stacked=False,
             figsize=(20, 10))  # pass a tuple (x, y) size

plt.title("Australia's 5 most adherent religions")
plt.ylabel('Number of religious adherents')
plt.xlabel('Years')

plt.show()
```

```python
df_top5.plot(kind='area', 
             alpha=0.25,  # 0 - 1, default value alpha = 0.5
             stacked=False,
             figsize=(20, 10))

plt.title("Australia's 5 most adherent religions")
plt.ylabel('Number of religious adherents')
plt.xlabel('Years')

plt.show()
```

*DUA TIPE DARI PLOTTING*

>Lapisan skrip (metode prosedural) 
-> menggunakan matplotlib.pyplot sebagai 'plt'



```python
df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10)) 
plt.title("Australia's 5 most adherent religions")
plt.ylabel('Number of religious adherents')
plt.xlabel('Years')
```

> Lapisan artist (Metode berorientasi objek)
 -> menggunakan instance Sumbu dari Matplotlib (lebih disukai)



```python
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title("Australia's 5 most adherent religions")
ax.set_ylabel('Number of religious adherents')
ax.set_xlabel('Years')
```

**BAR CHARTS**

```python
df_islam = df_can.loc['Islam', years]

df_islam.head()
```

```python
df_islam.plot(kind='bar', figsize=(25, 10), rot=90)

plt.xlabel('Year')
plt.ylabel('Number of Adherents')
plt.title('Histogram of Islam adherents in Australia from 2014 to 2016')

plt.show()
```

```python
df_islam.plot(kind='bar', figsize=(25, 10), rot=90)
plt.xlabel('Year')
plt.ylabel('Number of Adherents')
plt.title('Histogram of Islam adherents in Australia from 2014 to 2016')

# Annotate arrow
plt.annotate('', 
             xy=(125,270000), 
             xytext=(75,125), 
             xycoords='data',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

plt.show()
```
